```python
import numpy as np
```

```python
x = np.array([-2,3])
y = np.array([3,-4])
print(np.dot(x,y))
```

```python
def cosine_similarity(a,b):
  norm_a = np.linalg.norm(a)
  norm_b = np.linalg.norm(b)
  dot_product = np.dot(a,b)
  cosine_simlarity = dot_product/(norm_a*norm_b)
  return cosine_simlarity
```

```python
a = np.array([2,3])
b = np.array([3,4])

print(cosine_similarity(a,b))
```

```python
X = np.array([[1,2,3],
             [4,5,6],
             [1,0,0]
             ])
dot_products = X @ X.T

norms = np.linalg.norm(X,axis = 1)
norm_matrix = np.outer(norms,norms)
cosine_sim = dot_products/norm_matrix
print(cosine_sim)
```
